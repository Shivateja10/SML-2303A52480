```python
# QUESTION-1:
import pandas as pd
#LOAD YOUR EXCEL DATA
file_path='/content/Salary_Data.csv'
data=pd.read_csv(file_path)
#DISPLAY THE FIRST FEW ROWS OF THE DATA
print(data.head())
#ASSUMING THE LAST COLUMN IS THE TARGET VARIABLE AND OTHERS ARE FEATURES
x=data.iloc[:,:-1]#FEATURES
y=data.iloc[:,-1]#TARGET
#DISPLAY THE FEATURES AND TARGET
print("Features:\n",x.head())
print("Target:\n",y.head())
import matplotlib.pyplot as plt
for column in x.columns:
  #CHANGED X TO X
    plt.figure(figsize=(8,6))
    plt.scatter(x[column],y,alpha=0.5);
    #CHANGED X TO x
    plt.title(f'{column} vs target')
    plt.xlabel(column)
    plt.ylabel('Target')
    plt.show()
    from sklearn.preprocessing import MinMaxScaler
    scaler=MinMaxScaler()
    #NORMALIZE THE FEATURES AND TARGET
    X_normalized=scaler.fit_transform(x)
    y_normalized=scaler.fit_transform(y.values.reshape(-1,1))
    #COVERT THE BACK TO DATAFRAME FOR EASIER HANDLING
    X_normalized=pd.DataFrame(X_normalized,columns=x.columns)
    y_normalized=pd.Series(y_normalized.flatten(),name='Target')
    #DISPLAY THE NORMALIZED DATA
    print("Normalized Features:\n",X_normalized.head())

```
